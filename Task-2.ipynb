# Import Libraries

```python
# Import necessary libraries

from sklearn import datasets  # To load the Iris dataset
import pandas as pd  # For handling data in DataFrame
from sklearn.tree import DecisionTreeClassifier  # To build the Decision Tree model

from io import StringIO  # For handling string data
from IPython.display import Image  # To display images in Jupyter notebooks
from sklearn.tree import export_graphviz  # To export the tree in DOT format
import pydotplus  # To convert DOT format to a graph
```

# Load and Inspect Data

```python
iris = datasets.load_iris()  # Load the Iris dataset
df = pd.DataFrame(iris.data, columns=iris.feature_names)  # Create a DataFrame with the features of the Iris dataset
y = iris.target  # The target variable
```

```python
# Display the first 10 rows of the DataFrame
df.head(10)
```

```python
df.shape # Display the shape of the DataFrame (rows, columns)
```

# Initialize and Train Model

```python
clf = DecisionTreeClassifier()  # Initialize the Decision Tree classifier
clf.fit(df,y)  # Train the classifier on the dataset
```

# Visualize the Decision Tree

## Visualization without Feature and Class Names

```python
# Create DOT data for visualization without feature and class names
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png())  # Display the graph as a PNG image
```

## Visualization with Feature and Class Names

```python
# Create DOT data for visualization with feature and class names
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, filled=True, rounded=True, special_characters=True, 
                feature_names = iris.feature_names, class_names = iris.target_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png()) 
```
